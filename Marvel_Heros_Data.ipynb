### Data Visualization

```python
# Fill missing values in 'Alignment' with 'unknown'
filtered_heros = marvel_heros[marvel_heros['Alignment'] != 'unknown']
sns.set_theme(style="darkgrid")
plt.figure(figsize=(6, 4))
sns.countplot(x="Alignment",data = filtered_heros, hue = "Alignment", palette="pastel", legend = False)
plt.title("Distribution of Alignment")
plt.xlabel("Alignment")
plt.ylabel("Count")
plt.show()
```

```python
#Making a Bar plot for Maximum Power over Total for GOOD_Heros
plt.figure(figsize = (8,5))
sns.set_theme(style="darkgrid")
plt.bar(list(good_max_power['Name']),list(good_max_power['Total']),color = "b", label = 'Total_power')
plt.title('Maximum Power with Total',fontsize = 16, fontweight = "bold")
plt.xlabel('Name')
plt.ylabel('Total')
plt.legend(loc='upper left', fontsize=10, title='Group Legend',bbox_to_anchor=(1, 1))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set theme and figure size
sns.set_theme(style="darkgrid")
plt.figure(figsize=(8, 5))

# Create bar plot using Seaborn
ax = sns.barplot(x=good_max_power['Name'][0:5], 
                 y=good_max_power['Total'][0:5], 
                 color="b", 
                 label='Total_power')

# Customize the plot
plt.title('Maximum Power with Total', fontsize=16, fontweight="bold")
plt.xlabel('Name')
plt.ylabel('Total')

# Add legend with custom position
plt.legend(loc='upper left', fontsize=10, title='Group Legend', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
#Making a Bar plot for Maximum Power over Total for BAD_Heros
plt.figure(figsize = (8,5))
sns.set_theme(style="darkgrid")
plt.bar(list(bad_max_power['Name'])[0:5],list(bad_max_power['Total'])[0:5],color = "r", label = 'Total_power')
plt.title('Maximum Power with Total',fontsize = 16, fontweight = "bold")
plt.xlabel('Name')
plt.ylabel('Total')
plt.legend(loc='upper left', fontsize=10, title='Group Legend',bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Distribution of total power levels
plt.figure(figsize=(8, 5))
sns.histplot(marvel_heros["Total"], bins=20, kde=True, color="purple")
plt.title("Distribution of Total Power Levels")
plt.xlabel("Total Power Level")
plt.ylabel("Count")
plt.show()
```

```python
#Making a histogram fro speed for good superheroes
plt.figure(figsize=(5,5))
sns.set_theme(style="darkgrid")
plt.hist(good['Speed'],label = 'Speed')
plt.title('Distribution of Speed', fontsize = 10)
plt.xlabel('speed')
plt.legend()
plt.show()
```

```python
# Making a histogram for combat fro Good_Heros
plt.figure(figsize = (5,5))
sns.set_theme(style="darkgrid")
plt.hist(good['Combat'], label = 'Combat')
plt.title('Distribution of Combat', fontsize = 10)
plt.xlabel('Combat')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(marvel_heros.iloc[:, 2:].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Attributes")
plt.show()
```

```python
# Top 10 strongest superheroes
top_strongest = good.nlargest(10, "Strength")

plt.figure(figsize=(10, 5))
sns.barplot(x=top_strongest["Strength"], y=top_strongest["Name"], palette = "Reds_r")
plt.title("Top 10 Strongest Superheroes")
plt.xlabel("Strength")
plt.ylabel("Superhero")
plt.show()
```

```python
# Boxplot of Intelligence by Alignment
filtered_heros = marvel_heros[marvel_heros['Alignment'] != 'unknown']

plt.figure(figsize=(8, 5))
sns.boxplot(x=filtered_heros["Alignment"], y=filtered_heros["Intelligence"], palette="coolwarm")
plt.title("Distribution of Intelligence by Alignment")
plt.xlabel("Alignment")
plt.ylabel("Intelligence")
plt.show()
```
