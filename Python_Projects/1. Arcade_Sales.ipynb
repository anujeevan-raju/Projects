```python
#changing the data type of amount column
df['Amount'] = df['Amount'].astype('int')
```

# Data Visualization

```python
#Dividing customers based on Gender 
plt.figure(figsize=(8,5))
labels = ['Female','Male']
sns.set_style("whitegrid")
ax = sns.countplot(x="Gender", data=df, palette="coolwarm", hue = 'Gender', legend = False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender Distribution")
plt.legend(labels = labels, loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize = (8,5))
labels =['Femare','Male']
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', palette = 'Paired', legend = False)
for bars in ax.containers:
    ax.bar_label(bars)
plt.legend(labels = labels, loc = 'upper right')
plt.show()
```

```python
#Age Group Distribution
plt.figure(figsize=(8,5))
sns.set_style("whitegrid")
sns.countplot(x="Age Group", data=df, palette="viridis", hue = 'Age Group', legend = False)
plt.title('Age Group Distribution')
plt.show()
```

```python
# States with higher sales
plt.figure(figsize = (8,5))
state_sales = df.groupby("State")["Amount"].sum().sort_values(ascending=False).head(10)
state_sales.plot(kind="bar", colormap="plasma")
plt.xticks(rotation = 45)
plt.title("Top 10 States by Sales")
plt.show()
```

```python
plt.figure(figsize = (8,5))
labels = ['Occupation']
occupation_sales = df.groupby("Occupation")["Amount"].sum().sort_values(ascending=False)
occupation_sales.plot(kind="bar", colormap="cividis")
plt.title("Sales by Occupation")
plt.legend( labels = labels, loc = 'upper right')
plt.show()
```

```python
#Top category  by sales
plt.figure(figsize = (8,5))
product_sales = df.groupby("Product_Category")["Amount"].sum().sort_values(ascending=False).head(10)
product_sales.plot(kind="bar", colormap="magma")
plt.title("Top 10 Product Categories by Sales")
plt.show()
```

```python
#number of sales vs number orders
plt.figure(figsize =(8,5))
sns.scatterplot(x="Orders", y="Amount", data=df, hue = 'Amount', legend = False)
plt.title("Orders vs Amount")
plt.show()
```
